```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

login = {
    "Prince": {"username": 'prince', "matno": 1, "profile": 'CSCimgs/Prince.jpg', "Full Name": 'Prince Ibekwe'},
    "Maryam": {"username": 'maryam', "matno": 2, "profile": 'CSCimgs/Maryam.jpg', "Full Name": 'Maryam Shefiu'},
    "Chima": {"username": 'chima', "matno": 3, "profile": 'CSCimgs/Chima.jpg', "Full Name": 'Chima Okuwouke'},
    "Tomi": {"username": 'tomi', "matno": 4, "profile": 'CSCimgs/Tomi.jpg', "Full Name": 'Tomi Olasoji'},
    "Miracle": {"username": 'miracle', "matno": 5, "profile": 'CSCimgs/Miracle.jpg', "Full Name": 'Miracle Nnadiukwu'},
    "Bonaventure": {"username": 'bonaventure', "matno": 6, "profile": 'CSCimgs/Bonaventure.jpg', "Full Name": 'Bonaventure Mbbah-Offor'},
    "Zuky": {"username": 'zuky', "matno": 7, "profile": 'CSCimgs/Zuky.jpg', "Full Name": 'Chizuroke Onyekonwu'},
    "Oyinda": {"username": 'oyinda', "matno": 8, "profile": 'CSCimgs/Oyinda.jpg', "Full Name": 'Oyinda Apampa'},
    "Tabitha": {"username": 'tabitha', "matno": 9, "profile": 'CSCimgs/Tabitha.jpg', "Full Name": 'Tabitha Udezi'},
    "Love": {"username": 'love', "matno": 10, "profile": 'CSCimgs/Love.jpg', "Full Name": 'Love Folami'}
}

username = input('Please enter your username: ')
matno = int(input('Please input your matno: '))
option = int(input(
        "What Image technique would you like to use?\n"
        "1. Color Separation\n"
        "2. Brightness & Contrast\n"
        "3. Sharpening\n"
        "4. Blurring\n"
        "5. Scaling & Resizing\n"
        "6. Inversion\n"
        "Enter your choice (1-6): "
    ))

#def portraits(member):
#    image_path = login[member]["profile"]
#    image =cv2.imread(image_path)
#    cv2.imshow("Profile Picture", image)
#    cv2.waitKey(0)
#    cv2.destroyAllWindows()
    
#    return image
    
#found = False        
for member, details in login.items():
    if details["username"] == username and details["matno"] == matno:
        print(f"Login Successful! Welcome {details['Full Name']}!")
        #image = portraits(member)
        #found = True
        
        image_path = login[member]["profile"]
        image =cv2.imread(image_path)
             
        if option == 1:  # Color Separation
            B, G, R = cv2.split(image)
            cv2.imshow("Original", image)
            cv2.waitKey(0)
            cv2.imshow("Blue Channel", B)
            cv2.waitKey(0)
            cv2.imshow("Green Channel", G)
            cv2.waitKey(0)
            cv2.imshow("Red Channel", R)
            cv2.waitKey(0)
            cv2.destroyAllWindows()

        elif option == 2:
            plt.subplot(1, 2, 1)
            plt.title("Original")
            plt.imshow(image)
             
            brightness = 5
            contrast = 1.5
            image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)
             
            cv2.imwrite('img/contrast_image.jpg', image2)

            plt.subplot(1, 2, 2)
            plt.title("Brightness & Contrast")
            plt.imshow(image2)
            plt.show()
             
        elif option == 3:
            plt.subplot(1, 2, 1)
            plt.title("Original")
            plt.imshow(image)

            kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
    
            sharpened_image = cv2.filter2D(image, -1, kernel)

            cv2.imwrite('img/sharpened_image.jpg', sharpened_image)

            plt.subplot(1, 2, 2)
            plt.title("Sharpening")
            plt.imshow(sharpened_image)
            plt.show()
             
        elif option == 4: 
            plt.subplot(1, 2, 1)
            plt.title("Original")
            plt.imshow(image)

            filtered_image = cv2.medianBlur(image, 5)

            cv2.imwrite('img/Median-Blur.jpg', filtered_image)

            plt.subplot(1, 2, 2)
            plt.title("Median Blur")
            plt.imshow(filtered_image)
            plt.show()
             
        elif option == 5:
            plt.subplot(1, 2, 1)
            plt.title("Original")
            plt.imshow(image)

            scaled_image = cv2.resize(image, None, fx=2, fy=2)

            cv2.imwrite('img/Scaled.jpg', scaled_image)

            plt.subplot(1, 2, 2)
            plt.title("Scaled")
            plt.imshow(scaled_image)
            plt.show()
             
        elif option == 6:
            plt.subplot(1, 2, 1)
            plt.title("Original")
            plt.imshow(image)

            inverse_image = 255 - image

            cv2.imwrite('img/inverse_image.jpg', inverse_image)

            plt.subplot(1, 2, 2)
            plt.title("Inverse Color")
            plt.imshow(inverse_image)
            plt.show()
        break
            
    #else:
     #   print("Invalid username or password. Please try again.")
```
